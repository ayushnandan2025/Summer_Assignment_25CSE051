```python
%%writefile libmn.c
#include <stdio.h>
struct book {
  int id;
  char title[100];
  char author[50];
};
int main(){
  struct book b;
  printf("Enter book id: ");
  scanf("%d", &b.id);
  printf("Enter book title: ");
  scanf("%s", b.title);
  printf("Enter book author: ");
  scanf("%s", b.author);
  printf("\n---Book Details---");
  printf("Book id: %d\n", b.id);
  printf("Book title: %s\n", b.title);
  printf("Book author: %s\n", b.author);
  return 0;
}
```

```python
%%writefile bank.c
#include <stdio.h>
int main(){
  char name[50];
  int accno;
  float balance,deposit,withdraw;
  printf("Enter name: ");
  scanf("%s", name);
  printf("Enter account number: ");
  scanf("%d", &accno);
  printf("Enter initial balance: ");
  scanf("%f", &balance);
  printf("Enter deposit amount: ");
  scanf("%f", &deposit);
  balance += deposit;
  printf("Enter withdraw amount: ");
  scanf("%f", &withdraw);
  if(withdraw<=balance)
    balance -= withdraw;

  else
    printf("Insufficient balance\n");

  printf("\n---Account Details---");
  printf("Name: %s\n", name);
  printf("Account number: %d\n", accno);
  printf("balance:%.2f\n",balance);
  return 0;
}
```
