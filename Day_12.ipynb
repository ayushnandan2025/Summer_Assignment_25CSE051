```python
%%writefile pali.c
#include<stdio.h>
int ispalindrome(int n){
  int original=n,rev=0,digit;
  while(n!=0){
    digit=n%10;
    rev=rev*10+digit;
    n/=10;
  }
  return original==rev;
  }
  int main(){
    int n;
    printf("enter a number:");
    scanf("%d",&n);
    if(ispalindrome(n))
      printf("palindrome no");

    else
      printf("not a palindrome no");
    return 0;
  }
```

```python
%%writefile arms.c
#include <stdio.h>
int isarms(int n){
  int original=n,sum=0,rem;
  while(n!=0){
    rem=n%10;
    sum+=rem*rem*rem;
    n/=10;
  }
  return original==sum;
}
int main(){
  int n;
  printf("enter a number:");
  scanf("%d",&n);
  if(isarms(n))
    printf("arms no");
  else
    printf("not a arms no");
  return 0;
}
```

```python
%%writefile fibo.c
#include<stdio.h>
void fibo(int n){
  int a=0,b=1,c;
  for(int i=1;i<n;i++){
    printf("%d ",a);
    c=a+b;
    a=b;
    b=c;
  }
}
int main(){
  int n;
  printf("enter a number:");
  scanf("%d",&n);
  fibo(n);
  return 0;
}
```

```python
%%writefile perf.c
#include<stdio.h>
int isperf(int n){
  int sum=0;
  for(int i=1;i<n;i++){
    if(n%i==0)
      sum+=i;
  }
  return sum==n;
}
int main(){
  int n;
  printf("enter a number:");
  scanf("%d",&n);
  if(isperf(n))
    printf("perfect no");
  else
    printf("not a perfect no");
  return 0;
}
```
