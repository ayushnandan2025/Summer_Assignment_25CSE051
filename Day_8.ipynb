```python
%%writefile halfpyr.c
#include <stdio.h>
int main(){
  int rows;
  printf("Enter the number of rows: ");
  scanf("%d", &rows);
  for(int i=1; i<=rows; i++){
    for(int j=1; j<=i; j++){
      printf("*");
    }
    printf("\n");
  }
  return 0;
}
```

```python
%%writefile notri.c
#include <stdio.h>
int main(){
  int rows;
  printf("Enter the number of rows: ");
  scanf("%d", &rows);
  for(int i=1; i<=rows; i++){
    for(int j=1; j<=rows-i; j++)
      printf("%d",j);
      printf("\n");
    }
    return 0;
}
```

```python
!gcc notri.c
```

```python
!./a.out
```

```python
%%writefile chartri.c
#include <stdio.h>
int main(){
  int rows;
  printf("Enter the number of rows: ");
  scanf("%d", &rows);
  for(int i=1; i<=rows; i++){
    for(char ch='A';ch<='A'+i;ch++)
      printf("%c",ch);
    printf("\n");
}

return 0;
}
```
