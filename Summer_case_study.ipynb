```python
plt.figure(figsize=(7,7))
plt.pie(sales_df['Sales'],labels=sales_df['Product'],autopct='%1.1f%%')
plt.title("Sales Distribution")
plt.show()
```

```python
#Case Study 5
```

```python
visits=df['NumWebVisitsMonth']

print("Average Visits:", visits.mean())
print("Highest Visits:", visits.max())

# Categories
def visit_category(x):
    if x<=4:
      return "Low"
    elif x<=8:
       return "Medium"
    else:
        return "High"

df['Visit_Category'] =visits.apply(visit_category)
print(df['Visit_Category'].value_counts())

plt.figure(figsize=(7,4))
plt.hist(visits,bins=20)
plt.title("Website Visits Distribution")
plt.show()

sns.countplot(x='Visit_Category',data=df)
plt.title("Website Visits Category")
plt.show()
```

INSIGHT OF 5

Most customers visit the website around 5 times monthly

```python
country_count=df['Country'].value_counts()
print(country_count)


plt.figure(figsize=(8,5))
country_count.plot(kind='bar')
plt.title("Customer by Country")
plt.show()


avg_income=df.groupby('Country')[' Income '].mean().reset_index(name='Average Income')
print(avg_income)


df['TotalSpending']=(
    df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds']
)

avg_spending=df.groupby('Country')['TotalSpending'].mean()
print(avg_spending)


avg_spending.sort_values().plot(kind='bar')
plt.title("Average Spending by Country")
plt.show()
```

INSIGHT OF 7

Spain has the largest customer base.
