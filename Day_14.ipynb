```python
%%writefile linear.c
#include <stdio.h>
int main(){
  int n,key,found=0;
  printf("enter size of array");
  scanf("%d",&n);

  int arr[n];
  printf("enter array elements");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);

  printf("enter key element");
  scanf("%d",&key);

  for(int i=0;i<n;i++){
    if(arr[i]==key){
      printf("key element found at index %d",i+1);
      found=1;
      break;
    }
  }
  if(!found)
    printf("key element not found");
  return 0;
}
```

```python
%%writefile frequency.c
#include <stdio.h>
int main(){
  int n,key,count=0;
  printf("enter size of array");
  scanf("%d",&n);

  int arr[n];
  printf("enter array elements:\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);
    printf("enter element to find frequency:");
    scanf("%d",&key);

    for(int i=0;i<n;i++){
      if(arr[i]==key)
        count++;
    }
    printf("frequency of %d = %d",key,count);
    return 0;
}
```
