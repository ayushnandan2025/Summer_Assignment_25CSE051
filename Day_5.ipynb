```python
%%writefile perfect.c
#include<stdio.h>
int main(){
  int n,sum=0;
  printf("enter no:");
  scanf("%d",&n);
  for(int i=1;i<n;i++){
    if(n%i==0)
      sum=sum+i;
  }
  if(sum==n)
    printf("perfect no");
  else
    printf("not perfect no");
  return 0;
  }
```

```python
%%writefile strong.c
#include <stdio.h>
int main(){
  int n,sum=0,original,fact,rem;
  printf("enter no:");
  scanf("%d",&n);
  original=n;
  while(n>0){
    rem=n%10;
    int fact=1;
    for(int i=1;i<=rem;i++){
      fact=fact*i;
    }
    sum+=fact;
    n=n/10;
}
if(sum==original)
  printf("strong no");
else
  printf("not strong no");
return 0;
}
```

```python
!gcc strong.c
```

```python
!./a.out
```

```python
%%writefile factor.c
#include<stdio.h>
int main(){
  int n;
  printf("enter no:");
  scanf("%d",&n);
  printf("factors are:");
  for(int i=1;i<=n;i++){
    if(n%i==0)
      printf("%d ",i);
  }
  return 0;
}
```

```python
%%writefile lpf.c
#include <stdio.h>
int main(){
  int n,largest=0;

  printf("enter no:");
  scanf("%d",&n);
  for(int i=2;i<=n;i++){
    while(n%i==0){
      largest=i;
      n=n/i;
    }
  }
  printf("largest prime factor is %d",largest);
  return 0;
}


```
