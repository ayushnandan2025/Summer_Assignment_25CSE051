```python
%%writefile recufact.c
#include <stdio.h>
int factorial(int n){
  if(n==0 || n==1)
    return 1;
 return n*factorial(n-1);

}
int main(){
  int n;
  printf("enter no");
  scanf("%d",&n);
  printf("factorial of %d is %d",n,factorial(n));
  return 0;
}
```

```python
%%writefile recfib.c
#include <stdio.h>
int fib(int n){
  if(n==0)
    return 0;
  if(n==1)
    return 1;
  return fib(n-1)+fib(n-2);
}
int main(){
  int n;
  printf("enter no of terms");
  scanf("%d",&n);
  for(int i=0;i<n;i++)
    printf("%d ",fib(i));

  return 0;
}
```

```python
!./a.out
```

```python
%%writefile recsod.c
#include <stdio.h>

int sod(int n){
  if(n==0)
  return 0;
  return (n%10)+sod(n/10);
}
int main(){
  int n;
  printf("enter no");
  scanf("%d",&n);
  printf("sum of digits= %d",sod(n));
  return 0;
}
```

```python
%%writefile recrev.c
#include <stdio.h>
int reverse(int n,int rev){
  if(n==0)
    return rev;
  return reverse(n/10,rev*10+(n%10));
}
int main(){
  int n;
  printf("enter no");
  scanf("%d",&n);
  printf("reverse no=%d",reverse(n,0));
  return 0;
}
```
