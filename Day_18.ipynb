```python
%%writefile bubble.c
#include <stdio.h>
int main() {
  int n,temp;

  printf("enter size of array");
  scanf("%d",&n);
  int arr[n];
  printf("enter elements of array");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);
  for(int i=0;i<n-1;i++){
    for(int j=0;j<n-i-1;j++){
      if(arr[j]>arr[j+1]){
        temp=arr[j];
        arr[j]=arr[j+1];
        arr[j+1]=temp;
  }
    }
}
printf("sorted array:\n");
for(int i=0;i<n;i++)
  printf("%d ",arr[i]);
  return 0;
}
```

```python
%%writefile selec.c
#include <stdio.h>
int main(){
  int n,minind,temp;
  printf("enter size of array");
  scanf("%d",&n);
  int arr[n];
  printf("enter array elements:\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);
    for(int i=0;i<n-1;i++){
    minind=i;
    for(int j=i+1;j<n;j++){
      if(arr[j]<arr[minind])
        minind=j;

}
      temp=arr[i];
      arr[i]=arr[minind];
      arr[minind]=temp;
    }
    printf("sorted array:\n");
    for(int i=0;i<n;i++)
      printf("%d ",arr[i]);
      return 0;
}
```

```python
%%writefile bin.c
#include <stdio.h>
int main(){
  int n,key,low=0,high,mid;
  printf("enter size of array");
  scanf("%d",&n);
  int arr[n];
  printf("enter array elements:\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);
    printf("enter key element:");
    scanf("%d",&key);
    high=n-1;
    while(low<=high){
      mid=(low+high)/2;
      if(arr[mid]==key){
        printf("element found at index %d",mid+1);
        return 0;
      }
      else if(arr[mid]<key)
        low=mid+1;
      else
        high=mid-1;
    }

    return 0;
}
```

```python
%%writefile ardes.c
#include <stdio.h>
int main() {
  int n,temp;

  printf("enter size of array");
  scanf("%d",&n);
  int arr[n];
  printf("enter elements of array:\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);
    for(int i=0;i<n-1;i++){
      for(int j=i+1;j<n;j++){
        if(arr[i]<arr[j]){
          temp=arr[i];
          arr[i]=arr[j];
          arr[j]=temp;
        }
      }
    }
    printf("array in descending order:\n");
    for(int i=0;i<n;i++)
      printf("%d ",arr[i]);
  return 0;
}
```
