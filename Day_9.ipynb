```python
%%writefile revstar.c
#include <stdio.h>
int main(){
  int rows;
  printf("Enter the number of rows: ");
  scanf("%d", &rows);
  for (int i =rows; i >= 1; i--){
    for (int j = 1; j <= i; j++)
      printf("*");

    printf("\n");
  }
  return 0;
}
```

```python
%%writefile repchar.c
#include <stdio.h>
int main(){
  int rows;
  printf("Enter the number of rows: ");
  scanf("%d", &rows);
  for (int i =1; i <=rows ; i++){
    char ch='A'+i-1;
    for(int j=1; j<=i; j++)
      printf("%c",ch);

    printf("\n");
  }
  return 0;
}
```

```python
%%writefile holsq.c
#include <stdio.h>
int main(){
  int rows;
  printf("Enter size");
  scanf("%d",&rows);
  for (int i =1; i <=rows ; i++){
    for(int j=1; j<=rows; j++){
      if(i==1||i==rows||j==1||j==rows)
        printf("*");
      else
        printf(" ");
    }

    printf("\n");
  }
  return 0;
}
```
