```python
%%writefile prime.c
int main() {
  int n, i, flag = 1;
  printf("Enter a positive integer: ");
  scanf("%d", &n);
  if(n<=1)
  flag=0;
  for(i=2;i<=n/2;i++)
  {
    if(n%i==0)
    {
      flag=0;
      break;
    }
  }
  if(flag==1)
  printf("%d is a prime number",n);
  else
    printf("%d is not a prime number",n);
  return 0;
}
```

```python
%%writefile pnir.c
#include <stdio.h>
int main(){
  int start,end,i,j,flag;
  printf("Enter the range: ");
  scanf("%d %d",&start,&end);
  for (i=start;i<=end;i++){
    if(i<=1) continue;
    flag=1;
    for(j=2;j<=i/2;j++){
      if(i%j==0){
        flag=0;
        break;
      }
    }
    if(flag==1) printf("%d ",i);
  }
  return 0;
}
```

```python
%%writefile gcd.c
#include <stdio.h>
int main(){
  int a,b,gcd;
  printf("Enter two numbers: ");
  scanf("%d %d",&a,&b);
  for(int i=1;i<=a && i<=b;i++){
    if(a%i==0 && b%i==0)
     gcd=i;
  }
  printf("GCD of %d and %d is %d",a,b,gcd);
}
```

```python
%%writefile lcm.c
#include <stdio.h>
int main(){
  int a,b,lcm;
  printf("Enter two numbers: ");
  scanf("%d %d",&a,&b);
  for(int i=1;i<=a && i<=b;i++){
    if(a%i==0 && b%i==0)
     lcm=(a*b)/i;
  }
  printf("LCM of %d and %d is %d",a,b,lcm);
}
```
