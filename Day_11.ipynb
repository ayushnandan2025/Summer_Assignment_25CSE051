```python
%%writefile prime.c
#include <stdio.h>
int isprime(int n){
  if(n<=1)
   return 0;
  for(int i=2;i<=n/2;i++){
    if(n%i==0)
     return 0;
  }
  return 1;
}
int main(){
  int n;
  printf("enter no:");
  scanf("%d",&n);
  if(isprime(n))
   printf("prime");
  else
   printf("not prime");
  return 0;
}
```

```python
!./a.out
```
