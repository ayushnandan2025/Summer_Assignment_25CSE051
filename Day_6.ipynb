```python
%%writefile dtob.c
#include <stdio.h>
int main(){
  int n,binary[32],i=0;

  printf("Enter a decimal number: ");
  scanf("%d",&n);

  while(n>0){
    binary[i]=n%2;
    n=n/2;
    i++;
}
printf("binary= ");
for(int j=i-1;j>=0;j--){
  printf("%d",binary[j]);
}
return 0;
}
```

```python
%%writefile btod.c
#include <stdio.h>
int main(){
  int binary,decimal=0,base=1,rem;
  printf("Enter a binary number: ");
  scanf("%d",&binary);
  while(binary>0){
    rem=binary%10;
    decimal+=rem*base;
    base*=2;
    binary/=10;
}

printf("Decimal= %d",decimal);
return 0;
}
```

```python
!./a.out
```

```python
%%writefile countset.c
#include <stdio.h>
int main(){
  int n,count=0;
  printf("enter no:");
  scanf("%d",&n);
  while(n>0){
    if(n%2==1)
      count++;
    n/=2;
  }
printf("no of set bits=%d",count);
return 0;
}
```

```python
%%writefile power.c
#include <stdio.h>
int main(){
  int x,n;
  long long result=1;
  printf("enter base and power");
  scanf("%d%d",&x,&n);

  for(int i=1;i<=n;i++){
    result*=x;
  }
  printf("result=%lld",result);
  return 0;
}
```
