```python
%%writefile fibonacci.c
#include <stdio.h>
int main(){
  int n,a=0,b=1,c;

  printf("enter no of terms: ");
  scanf("%d",&n);

  printf("fibonacci series: ");


  for(int i=1;i<n;i++){

    printf("%d ",a);
    c=a+b;
    a=b;
    b=c;
  }
  return 0;
}
```

```python
%%writefile armstrong.c
#include <stdio.h>
int main(){
  int n,original,rem,sum=0;

  printf("enter no:");
  scanf("%d",&n);
  original=n;

  while(n>0){
    rem=n%10;
    sum=sum+(rem*rem*rem);
    n=n/10;
  }
  if(sum==original)
    printf("armstrong no");
  else
    printf("not armstrong no");
    return 0;
  }

```

```python
%%writefile armrange.c
#include <stdio.h>
int main(){
  int start,end,num,temp,rem,sum;

  printf("enter range:");
  scanf("%d%d",&start,&end);
  printf("armstrong no:");
  for(num=start;num<=end;num++){

    temp = num;
    sum=0;
    while(temp>0){
      rem=temp%10;
      sum=sum+(rem*rem*rem);
      temp=temp/10;
    }
    if(sum==num)
      printf("%d ",num);
  }
  return 0;
}
```
