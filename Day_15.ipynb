```python
%%writefile reverse.c
#include <stdio.h>
int main(){
  int n;
  printf("enter size of array:");
  scanf("%d",&n);

  int arr[n];
  printf("enter array elements:\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);
  printf("reverse array is:\n");
  for(int i=n-1;i>=0;i--)
    printf("%d",arr[i]);
    return 0;
}
```

```python
%%writefile rotlef.c
#include <stdio.h>
int main(){
  int n,first;
  printf("enter size of array:");
  scanf("%d",&n);

  int arr[n];
  printf("enter array elements:\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);

first=arr[0];
for(int i=0;i<n-1;i++)
  arr[i]=arr[i+1];
arr[n-1]=first;
printf("Array after left rotation:\n");
for(int i=0;i<n;i++)
  printf("%d ",arr[i]);
  return 0;
}
```

```python
%%writefile rotrig.c
#include <stdio.h>
int main(){
  int n,last;
  printf("enter size of array:");
  scanf("%d",&n);

  int arr[n];
  printf("enter array elements:\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);

last=arr[n-1];
for(int i=n-1;i>0;i--)
  arr[i]=arr[i-1];
arr[0]=last;
printf("Array after right rotation:\n");
for(int i=0;i<n;i++)
  printf("%d ",arr[i]);
  return 0;
}
```
