```python
%%writefile hello.c
#include<stdio.h>
int main()
{
int n,sum=0;
printf("enter n:");
scanf("%d",&n);

for(int i=1; i<=n; i++){
sum+=i;
}
printf("Sum=%d",sum);
return 0;
}
```

```python
%%writefile multiplication_table.c
#include <stdio.h>
int main(){
    int n;
    printf("Enter no:");
    scanf("%d",&n);

    for(int i=1; i<=10; i++){
        printf("%d × %d =%d\n",n,i,n*i);
    }
    return 0;
}
```

```python
!gcc multiplication_table.c
```

```python
!./a.out
```

```python
%%writefile factorial.c
#include <stdio.h>
int main(){
    int n,fact=1;
    printf("Enter no:");
    scanf("%d",&n);

    for(int i=1; i<=n; i++){
        fact*=i;
    }
    printf("Factorial=%d",fact);
}


```

```python
!gcc factorial.c
```

```python
!./a.out
```

```python
%%writefile count.c
#include<stdio.h>
int main(){
  int n,count=0;
  printf("enter no:");
  scanf("%d",&n);

  while(n!=0){
    n/=10;
    count++;

  }
  printf("digits =%d",count);
  return 0;
}
```

```python
!gcc count.c
```
