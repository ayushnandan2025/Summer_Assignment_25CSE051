```python
%%writefile pal.c
#include <stdio.h>
int main(){
  char str[100];
  int i,len=0,flag=1;
  printf("enter a string");
  fgets(str,sizeof(str),stdin);
  while(str[len]!='\0' && str[len]!='\n')
    len++;

  for(i=0;i<len/2;i++){
    if(str[i]!=str[len-i-1]){
      flag=0;
      break;
    }
}
if (flag==1)
printf("palindrome string");
else
printf("not palindrome string");
return 0;
}
```

```python
%%writefile charfre.c
#include <stdio.h>
int main(){
  char str[100],ch;
  int count=0;

  printf("enter a string:");
  fgets(str,sizeof(str),stdin);
  printf("enter a character to search:");
  scanf("%c",&ch);
  for(int i=0;str[i]!='\0';i++){
    if(str[i]==ch)
      count++;
  }
  printf("frequency of %c=%d",ch,count);
  return 0;
}
```
