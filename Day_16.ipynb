```python
%%writefile misno.c
#include <stdio.h>
int main(){
  int n,sum=0,expsum;
  printf("enter value of n");
  scanf("%d",&n);
  int arr[n-1];
  printf("enter %d elements:\n",n-1);
  for(int i=0;i<n-1;i++){
    scanf("%d",&arr[i]);
    sum=sum+arr[i];
  }
  expsum=n*(n+1)/2;
  printf("missing number = %d",expsum-sum);
  return 0;

}
```

```python
%%writefile maxfre.c
#include <stdio.h>
int main(){
  int n;
  printf("enter size of array");
  scanf("%d",&n);
  int arr[n];
  printf("enter array elements:\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);
  int maxfre=0,element;
  for(int i=0;i<n;i++){
    int count=1;
    for(int j=i+1;j<n;j++){
      if(arr[i]==arr[j])
        count++;

    }
    if(count>maxfre){
      maxfre=count;
      element=arr[i];
    }
  }
  printf("element with maximum frequency = %d",element);
  return 0;
}
```

```python
%%writefile pairsum.c
#include <stdio.h>
int main(){
  int n,target,found=0;
  printf("enter size of array");
  scanf("%d",&n);
  int arr[n];
  printf("enter array elements:\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);
    printf("enter target sum");
  scanf("%d",&target);
  printf("pairs are:\n");
  for(int i=0;i<n-1;i++){
    for(int j=i+1;j<n;j++){
      if(arr[i]+arr[j]==target){
        printf("(%d %d)\n",arr[i],arr[j]);
        found=1;
      }
    }
  }
  if(!found)
    printf("no pair found");
  return 0;
}
```
