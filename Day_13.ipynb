```python
%%writefile array.c
#include<stdio.h>
int main(){
  int n;
  printf("enter size of an array");
  scanf("%d",&n);


  int arr[n];
  printf("enter %d elements :\n",n);
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);

  printf("array elements are:\n");
  for(int i=0;i<n;i++)
    printf("%d\n",arr[i]);

    return 0;
}
```

```python
%%writefile sanda.c
#include <stdio.h>
int main(){
  int n,sum=0;
  printf("enter size of an array");
  scanf("%d",&n);


  int arr[n];
  printf("enter elements :\n");
  for(int i=0;i<n;i++){
    scanf("%d",&arr[i]);
    sum+=arr[i];
}

printf("sum=%d\n",sum);
printf("avg=%.2f",(float)sum/n);
return 0;
}
```

```python
%%writefile larsma.c
#include <stdio.h>
int main(){
  int n;
  printf("enter size of array:");
  scanf("%d",&n);
  int arr[n];
  printf("enter elements:\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);

    int largest=arr[0];
    int smallest=arr[0];
    for(int i=1;i<n;i++){
      if(arr[i]>largest)
        largest=arr[i];
        if(arr[i]<smallest)
        smallest=arr[i];
    }
    printf("largest=%d\n",largest);
    printf("smallest=%d\n",smallest);
    return 0;
}
```

```python
%%writefile evod.c
#include <stdio.h>
int main(){
  int n,even=0,odd=0;
  printf("enter size of an array");
  scanf("%d",&n);
  int arr[n];
  printf("enter elements :\n");
  for(int i=0;i<n;i++)
    scanf("%d",&arr[i]);
  for(int i=0;i<n;i++){
    if(arr[i]%2==0)
      even++;
    else
      odd++;
  }
  printf("even=%d\n",even);
  printf("odd=%d\n",odd);
  return 0;
}
```
