```python
%%writefile stlen.c
#include <stdio.h>
int main(){
  char str[100];
  int length=0;
  printf("enter a string:");
  fgets(str,sizeof(str),stdin);
  while(str[length]!='0'&&str[length]!='\n')
    length++;

  printf("length of string = %d",length);
  return 0;
}
```

```python
%%writefile revst.c
#include<stdio.h>
int main(){
  char str[100],temp;
  int i,length=0;
  printf("enter a string:");
  fgets(str,sizeof(str),stdin);
  while(str[length]!='0'&&str[length]!='\n')
    length++;
    for(i=0;i<length/2;i++){
      temp=str[i];
      str[i]=str[length-i-1];
      str[length-i-1]=temp;
    }
    printf("reverse string = %s",str);
    return 0;
}
```
