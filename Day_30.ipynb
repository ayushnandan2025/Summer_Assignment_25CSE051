```python
%%writefile hms.c
#include <stdio.h>
struct patient{
  int id;
  char name[20];
  int age;
  char disease[50];
};
int main(){
  struct patient p;
  printf("Enter patient id: ");
  scanf("%d", &p.id);
  printf("Enter patient name: ");
  scanf("%s", p.name);
  printf("Enter patient age: ");
  scanf("%d", &p.age);
  printf("Enter patient disease: ");
  scanf("%s", p.disease);
  printf("\n---Patient Details---\n");
  printf("Patient id: %d\n", p.id);
  printf("Patient name: %s\n", p.name);
  printf("Patient age: %d\n", p.age);
  printf("Patient disease: %s\n", p.disease);
  return 0;

}
```

```python
%%writefile hmhs.c
#include <stdio.h>
struct cust{
  int roomno;
  char name[50];
  int days;
  float rpd;
};
int main(){
  struct cust c;
  float total;
  printf("Enter room number: ");
  scanf("%d", &c.roomno);
  printf("Enter customer name: ");
  scanf("%s", c.name);
  printf("Enter number of days: ");
  scanf("%d", &c.days);
  printf("Enter room per day price: ");
  scanf("%f", &c.rpd);
  total=c.days*c.rpd;
  printf("\n---Customer Details---\n");
  printf("Room number: %d\n", c.roomno);
  printf("Customer name: %s\n", c.name);
  printf("Number of days: %d\n", c.days);
  printf("Room per day price: %.2f\n", c.rpd);
  printf("Total amount: %.2f\n", total);
  return 0;

}
```

```python
%%writefile sbs.c
#include <stdio.h>
int main(){
  char item[50];
  int qty;
  float price, total;
  printf("Enter item name: ");
  scanf("%s", item);
  printf("Enter quantity: ");
  scanf("%d", &qty);
  printf("Enter price per item: ");
  scanf("%f", &price);
  total=qty*price;
  printf("\n---Bill---\n");
  printf("Item name: %s\n", item);
  printf("Quantity: %d\n", qty);
  printf("Price per item: %.2f\n", price);
  printf("Total amount: %.2f\n", total);
  return 0;

}
```

```python
%%writefile srms.c
#include <stdio.h>
struct student{
  int rollno;
  char name[50];
  int m1,m2,m3;
  int total;
  float per;
};
int main(){
  struct student s;
  printf("Enter roll number: ");
  scanf("%d", &s.rollno);
  printf("Enter student name: ");
  scanf("%s", s.name);
  printf("Enter marks for 3 subjects: ");
  scanf("%d%d%d", &s.m1,&s.m2,&s.m3 );
  s.total=s.m1+s.m2+s.m3;
  s.per=s.total/3.0;
  printf("\n---Student Details---\n");
  printf("Roll number: %d\n", s.rollno);
  printf("Student name: %s\n", s.name);
  printf("marks       :%d%d%d\n",s.m1 ,s.m2 ,s.m3);
  printf("Total marks: %d\n", s.total);
  printf("Percentage: %.2f\n", s.per);
  if(s.per>=90)
    printf("Grade: A\n");
  else if(s.per>=80)
    printf("Grade: B\n");
  else if(s.per>=70)
    printf("Grade: C\n");
  else if(s.per>=60)
    printf("Grade: D\n");
  else if(s.per>=50)
    printf("Grade: E\n");
  else
    printf("Grade: F\n");
  return 0;


}
```
