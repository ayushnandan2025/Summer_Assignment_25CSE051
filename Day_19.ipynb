```python
%%writefile admat.c
#include <stdio.h>
int main(){
  int r,c;

  printf("enter rows and columns");
  scanf("%d%d",&r,&c);
  int a[r][c],b[r][c],sum[r][c];

  printf("enter first matrix:\n");
  for(int i=0;i<r;i++)
    for(int j=0;j<c;j++)
      scanf("%d",&a[i][j]);

      printf("enter second matrix:\n");
  for(int i=0;i<r;i++)
    for(int j=0;j<c;j++)
      scanf("%d",&b[i][j]);

      printf("sum of two matrices:\n");
  for(int i=0;i<r;i++){
    for(int j=0;j<c;j++){
      sum[i][j]=a[i][j]+b[i][j];
      printf("%d\t",sum[i][j]);
}
printf("\n");
  }
  return 0;





}
```

```python
%%writefile submat.c
#include <stdio.h>
int main(){
  int r,c;
  printf("enter rows and columns");
  scanf("%d%d",&r,&c);
  int a[r][c],b[r][c],sub[r][c];

  printf("enter first matrix:\n");
  for(int i=0;i<r;i++)
    for(int j=0;j<c;j++)
      scanf("%d",&a[i][j]);

      printf("enter second matrix:\n");
      for(int i=0;i<r;i++)
    for(int j=0;j<c;j++)
      scanf("%d",&b[i][j]);

      printf("subtract matrix:\n");
      for(int i=0;i<r;i++){
    for(int j=0;j<c;j++){
      sub[i][j]=a[i][j]-b[i][j];
      printf("%d\t",sub[i][j]);
}
printf("\n");
  }
  return 0;
}
```
