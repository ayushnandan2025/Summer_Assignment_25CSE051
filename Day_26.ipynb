```python
%%writefile nogue.c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int main(){
  int guess,no;
  srand(time(0));
  no=rand() % 100+1;
  printf("Guess a number between 1 and 100\n");
  do{
    printf("Enter your guess: ");
    scanf("%d",&guess);
    if(guess>no)
      printf("Too high\n");

    else if(guess<no)
      printf("Too low\n");
    else
      printf("Correct\n");

   } while(guess!=no);
   return 0;

}
```

```python
%%writefile atm.c
#include <stdio.h>
int main(){
  int choice;
  float balance=10000;
  float amount;
  do{
    printf("\n---ATM MENU---");
    printf("1. Check balance\n");
    printf("2. Deposit\n");
    printf("3. Withdraw\n");
    printf("4. Exit\n");
    printf("Enter your choice: ");
    scanf("%d",&choice);
    switch(choice){
      case 1:
        printf("Your balance is %.2f\n",balance);
        break;
        case 2:
        printf("Enter amount to deposit: ");
        scanf("%f",&amount);
        balance+=amount;
        printf("Amount deposited successfully\n");
        break;
        case 3:
        printf("Enter amount to withdraw: ");
        scanf("%f",&amount);
        if(amount>balance)
          printf("Insufficient balance\n");
        else{
          balance-=amount;
          printf("Amount withdrawn successfully\n");

  }
  break;
  case 4:
  printf("Thank you for using our ATM\n");
  break;
  default:
  printf("Invalid choice\n");
}
} while(choice!=4);
return 0;
}
```
