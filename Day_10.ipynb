```python
%%writefile stpy.c
#include <stdio.h>
int main(){
  int rows;
  printf("Enter number of rows: ");
  scanf("%d", &rows);
  for(int i = 1; i <= rows; i++){
    for(int j = 1; j <= rows-i; j++)
      printf(" ");
      for(int j=1; j<=(2*i-1); j++)
        printf("*");

    printf("\n");
  }
  return 0;
}
```

```python
%%writefile revpyr.c
#include <stdio.h>
int main(){
  int rows;
  printf("Enter number of rows: ");
  scanf("%d", &rows);
  for(int i =rows; i >= 1; i--){
    for(int j = 1; j <= rows-i; j++)
      printf(" ");
      for(int j=1; j<=(2*i-1); j++)
        printf("*");

    printf("\n");
  }
  return 0;
}
```

```python
%%writefile numpyr.c
#include <stdio.h>
int main(){
  int rows;
  printf("Enter number of rows: ");
  scanf("%d", &rows);
  for(int i =1; i <= rows; i++){
    for(int j = 1; j <= rows-i; j++)
      printf(" ");
      for(int j=1; j<=i; j++)
        printf(" ");
    for(int j=1; j<=i; j++)
      printf("%d",j);
      for(int j=i-1; j>=1; j--)
      printf("%d",j);

    printf("\n");
  }
  return 0;
}
```

```python
!gcc numpyr.c
```

```python
!./a.out
```

```python
%%writefile charpyr.c
#include <stdio.h>
int main(){
  int rows;
  printf("Enter number of rows: ");
  scanf("%d", &rows);
  for(int i =1; i <= rows; i++){
    for(int j = 1; j <= rows-i; j++)
      printf(" ");
      for(char ch='A'; ch<'A'+i; ch++)
        printf("%c",ch);
        for(char ch='A'+i-2; ch>='A'; ch--)
        printf("%c",ch);

    printf("\n");
  }
  return 0;
}
```
