```python
%%writefile merso.c
#include <stdio.h>
int main(){
  int n1,n2;
  printf("enter size of first array:");
  scanf("%d",&n1);
  int a[n1];
  printf("enter sorted element of first array:\n");
  for(int i=0;i<n1;i++)
    scanf("%d",&a[i]);
    printf("enter size of second array:");
  scanf("%d",&n2);
  int b[n2];
  printf("enter sorted element of second array:\n");
  for(int i=0;i<n2;i++)
    scanf("%d",&b[i]);
  int c[n1+n2];
  int i=0,j=0,k=0;
  while(i<n1 && j<n2){
    if(a[i]<b[j])
      c[k++]=a[i++];
      else
      c[k++]=b[j++];
  }
  while(i<n1)
    c[k++]=a[i++];
  while(j<n2)
    c[k++]=b[j++];
    printf("merged array is:\n");
    for(int i=0;i<n1+n2;i++)
      printf("%d ",c[i]);
return 0;


}
```

```python
%%writefile cchst.c
#include <stdio.h>
int main(){
  char str1[100],str2[100];
  printf("enter first string:");
  scanf("%s",str1);
  printf("enter second string:");
  scanf("%s",str2);
  printf("common char:");
  for(int i=0;str1[i]!='\0';i++){
    for(int j=0;str2[j]!='\0';j++){
      if(str1[i]==str2[j]){
        printf("%c",str1[i]);
        break;
}
    }
  }
  return 0;}
```

```python
%%writefile naal.c
#include <stdio.h>
#include <string.h>
int main(){
  int n;
  char names[20][50],temp[50];
  printf("enter number of names:");
  scanf("%d",&n);
  printf("enter names:\n");
  for(int i=0;i<n;i++)
    scanf("%s",names[i]);
    for(int i=0;i<n-1;i++){
    for(int j=i+1;j<n;j++){
      if(strcmp(names[i],names[j])>0){
        strcpy(temp,names[i]);
        strcpy(names[i],names[j]);
        strcpy(names[j],temp);
      }
    }
  }
  printf("names in dictionary order:\n");
  for(int i=0;i<n;i++)
    printf("%s\n",names[i]);
  return 0;
}
```

```python
%%writefile sole.c
#include <stdio.h>
#include <string.h>
int main(){
  char words[20][50],temp[50];
  int n;
  printf("enter number of words:");
  scanf("%d",&n);
  printf("enter words:\n");
  for(int i=0;i<n;i++)
    scanf("%s",words[i]);
    for(int i=0;i<n-1;i++){
    for(int j=i+1;j<n;j++){
      if(strlen(words[i])>strlen(words[j])){
        strcpy(temp,words[i]);
        strcpy(words[i],words[j]);
        strcpy(words[j],temp);
      }
    }

}
printf("words sorted by length:\n");
for(int i=0;i<n;i++)
  printf("%s\n",words[i]);
return 0;
}
```
