```python
%%writefile nonrep.c
#include <stdio.h>
int main(){
  char str[100];
  int i,j, unique;
  printf("enter a string");
  fgets(str,sizeof(str),stdin);
  for(i=0;str[i]!='\0'&&str[i]!='\n';i++){
    unique=1;
    for(j=0;str[j]!='\0' && str[j]!='\n';j++){
      if(i!=j && str[i]==str[j]){
        unique=0;
        break;
      }
    }
    if(unique){
      printf("first non repeating char=%c",str[i]);
      return 0;
    }
}
printf(" no non repeating char found");
return 0;}
```

```python
%%writefile firep.c
#include <stdio.h>
int main(){
  char str[100];
  printf("enter a string:");
  fgets(str,sizeof(str),stdin);
  for(int i=0;str[i]!='\0'&&str[i]!='\n';i++){
    for(int j=i+1;str[j]!='\0'&&str[j]!='\n';j++){
      if(str[i]==str[j]){
        printf("first repeating char=%c",str[i]);
        return 0;
      }
    }
}
printf("no repeating char found");
return 0;}
```

```python
%%writefile anagram.c
#include <stdio.h>
#include <string.h>
int main(){
  char str1[100],str2[100];
  int count1[256]={0},count2[256]={0};
  printf("enter first string:");
  scanf("%s",str1);
  printf("enter second string:");
  scanf("%s",str2);
  for(int i=0;str1[i]!='\0' ;i++)
  count1[(unsigned char)str1[i]]++;
  for(int i=0;str2[i]!='\0' ;i++)
  count2[(unsigned char)str2[i]]++;
  for(int i=0;i<256;i++){
    if(count1[i]!=count2[i]){
      printf("not anagram");
      return 0;
    }
  }
  printf("anagram");
return 0;
}
```

```python
%%writefile maxch.c
#include <stdio.h>
int main(){
  char str[100];
  int freq[256]={0};
  int max=0;
  char ch;
  printf("enter a string:");
  fgets(str,sizeof(str),stdin);
  for(int i=0;str[i]!='\0'&&str[i]!='\n';i++){
    freq[(unsigned char)str[i]]++;
    for(int i=0;i<256;i++){
      if(freq[i]>max){
        max=freq[i];
        ch=i;
      }
    }
  }
  printf("max repeating char=%c",ch);
  printf("\nfreq=%d",max);
  return 0;

}
```
